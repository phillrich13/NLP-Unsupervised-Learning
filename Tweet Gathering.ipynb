```python
import tweepy
import pandas as pd
import json
import csv
import re
import os
import requests
import time
import pickle
import logging
```

```python
logging.basicConfig(filename=r'/Users/philliprichardson/Metis/Module 5/twitter_error_log.log',filemode='w', level=logging.ERROR)
```

```python
path = r'/Users/philliprichardson/Metis/Module 5/keys.json'
def get_keys(path):
    with open(path) as f:
        return json.load(f)
    
keys = get_keys(path)

consumer_key=keys['API_key']
consumer_secret=keys['API_secret_key']
access_token=keys['access_token']
access_token_secret=keys['access_token_secret']

auth = tweepy.OAuthHandler(consumer_key,consumer_secret)
auth.set_access_token(access_token,access_token_secret)
```

```python
api = tweepy.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True, parser=tweepy.parsers.JSONParser())
```

```python
maxTweets = 1000000 # Some arbitrary large number
tweetsPerQry = 100  # this is the max the API permits

# If results from a specific ID onwards are reqd, set since_id to that ID.
# else default to no lower limit, go as far back as API allows
sinceId = None

# If results only below a specific ID are reqd, set max_id to that ID.
# else default to no upper limit, start from the most recent tweet matching the search query.
max_id = 0

tweetCount = 0
tweetFiles = 0
msg =  "Downloading max {0} tweets".format(maxTweets)
logging.error(msg)
tweets = []
while 4000*tweetFiles + tweetCount < maxTweets:
    try:
        if (max_id <= 0):
            if (not sinceId):
                new_tweets = api.search(q='e32021', count=tweetsPerQry, lang = 'en')
            else:
                new_tweets = api.search(q='e32021', count=tweetsPerQry,
                                        since_id=sinceId, lang = 'en')
        else:
            if (not sinceId):
                new_tweets = api.search(q='e32021', count=tweetsPerQry,
                                        max_id=str(max_id - 1), lang = 'en')
            else:
                new_tweets = api.search(q='e32021', count=tweetsPerQry,
                                        max_id=str(max_id - 1),
                                        since_id=sinceId, lang = 'en')
                
        new_tweets = new_tweets['statuses']
        if len(new_tweets) == 0:
            msg = 'No tweets found'
            logging.error(msg)
            break
        
        tweets.extend(new_tweets)
        tweetCount += len(new_tweets)
        max_id = new_tweets[-1]['id']
        
        if tweetCount > 4000:
            with open(r'/Users/philliprichardson/Metis/Module 5/Tweets/tweets'+str(tweetFiles)+'.json', 'w') as outfile:  
                json.dump(tweets, outfile)
            msg = "JSON file saved"
            logging.error(msg)
            tweetFiles += 1
            tweetCount = 0
            del tweets[:]
                
    except tweepy.TweepError as e:
        msg = 'Query failed when max_id equaled {0}: {1}'.format(max_id, e)
        logging.error(msg)

if tweetCount > 0:
    with open(r'/Users/philliprichardson/Metis/Module 5/Tweets/tweets'+str(tweetFiles)+'.json', 'w') as outfile:  
        json.dump(tweets, outfile)
    msg = "JSON file saved"
    logging.error(msg)
    tweetFiles += 1
    tweetCount = 0
```
