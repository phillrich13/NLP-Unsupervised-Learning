# MVP

I have downloaded ~122,000 tweets containing the official hashtag of the E3 conference (#E32021) using tweepy. These tweets were pulled on 6/15 between. 8p.m. and midnight (this matters because Nintendo's conference was on 6/15 so there will be less tweets for Nintendo compared to companies that went earlier). My first step was to pre-process the tweets. This included removing hyperlinks, '...', lowercase, splitting specific hashtags (each company had a specific hashtag for their presentation), removing stop words, removing punctuation, and more. I then fit 2 vectorizers for testing (TF-IDF and CountVectorizer), as well as 2 topic modelers (NMF, LSA) and then tested multiple iterations of the vectorizers and topic models. My best topic modeler (NMF with 7 topics and TF-IDF vectorizer) produced the following topics:
